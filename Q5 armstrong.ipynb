```python
def print_fibonacci(count):
    a, b = 0, 1
    sequence = []
    if count >= 1:
        sequence.append(a)
    if count >= 2:
        sequence.append(b)
        
    for _ in range(2, count):
        next_num = a + b
        sequence.append(next_num)
        a = b
        b = next_num
        
    print(f"Fibonacci Sequence ({count} terms): {sequence}")


print_fibonacci(10)
```
