```python
class Rational(object):
  def __init__(self,nume,deno):
    self.nume=nume
    self.deno=deno
    self._reduce()
  def numerator(self):
    return self.nume
  def denominator(self):
    return self.deno
  def __str__(self):
    return f"{self.nume}/{self.deno}"
  def _reduce(self):
    div=self._gcd(self.nume,self.deno)
    self.nume=self.nume//div
    self.deno=self.deno//div
  def _gcd(self,a,b):
    (a,b)=(max(a,b),min(a,b))
    while b>0:
      (a,b)=(b,a%b)
    return a
  def __add__(self,other):
    newnum=self.nume * other.deno + other.nume * self.deno
    newden=self.deno * other.deno
    return Rational(newnum,newden)
  def __sub__(self,other):
    newnum=self.nume * other.deno - other.nume * self.deno
    newden=self.deno * other.deno
    return Rational(newnum,newden)
  def __mul__(self, other):
    return Rational(self.nume * other.nume, self.deno * other.deno)
  def __eq__(self, other):
    return self.nume == other.nume and self.deno == other.deno
  def __lt__(self, other):
    return self.nume * other.deno < other.nume * self.deno
  def __gt__(self, other):
    return self.nume * other.deno > other.nume * self.deno
```

```python
r1=Rational(2,4)
r2=Rational(3,6)
r3=Rational(1,6)
print(r1+r3)
print(r1>r3)
print(r2<r3)
print(r1*r2)
print(r1-r3)
print(r1,r2)
print(r1==r2)
```
